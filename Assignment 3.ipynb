```python
# lgis
def getIncreasingSubs(n, perm):
    #Using Binary search 
    
    nums = inputHandler(perm)
    sequence = []

    sequence.append(nums[0])

    for i in range(1, n):
        if nums[i] > sequence[-1]:
            
            sequence.append(nums[i])
        else:
            low = 0
            high = len(sequence) - 1
            while low < high:
                mid = low + (high - low) // 2
                if sequence[mid] < nums[i]:
                    low = mid + 1
                else:
                    high = mid
            sequence[low] = nums[i]
    return sequence

def getDecreasingSubs(n, perm):
    
    nums = inputHandler(perm)
    sequence = []

    sequence.append(nums[0])

    for i in range(1, n):
        if nums[i] < sequence[-1]:
            sequence.append(nums[i])
        else:
            low = 0
            high = len(sequence) - 1
            while low < high:
                mid = low + (high - low) // 2
                if sequence[mid] > nums[i]:
                    low = mid + 1
                else:
                    high = mid
            sequence[low] = nums[i]
    return sequence

def inputHandler(perm):
    temp = perm.split()
    temp1 = map(int, temp)
    nums = list(temp1)
    return nums

def finalResult(n, perm):
    increasing = getIncreasingSubs(n, perm)
    decreasing = getDecreasingSubs(n, perm)
    incr = '' 
    decr = ''
    for i in increasing:
        incr = incr + str(i) + ' '
    for j in decreasing: 
        decr = decr + str(j) + ' '

    return incr + '\n' + decr

print(finalResult(9322, '6710 2712 8514 450 379 4368 2410 1571 8531 1060 8708 4184 335 262 6601 1056 4195 9197 3464 5134 4135 631 1139 9044 6411 8645 3120 1676 5355 1143 3157 9186 7324 5814 7080 9096 2717 8410 5350 7936 2451 1603 5209 8813 412 7908 4737 620 8951 3659 4847 1194 7269 6612 28 8137 4708 1858 2734 4706 7322 1083 2049 6093 5371 1758 6348 2790 4502 6621 7389 4845 4538 1020 7237 2644 2626 1418 5986 2207 1843 1439 4656 7583 3163 3510 9089 3213 5307 8002 2797 4451 5550 4399 6479 7551 634 8224 7199 8518 2269 4288 8288 3525 4211 8736 584 9280 2094 1324 4192 2575 3208 1840 9017 8788 5220 7781 4747 9021 4049 2106 9149 6602 8838 3077 6039 1462 8181 6444 7477 6609 8977 6220 7079 900 4780 272 1160 1772 8794 1408 5517 825 7447 3420 8750 135 5242 6284 4169 8182 628 6999 9161 8667 6886 7689 8452 388 3204 7740 7232 3757 7598 1421 9038 7107 4743 3399 7347 1120 2900 710 6707 7538 1126 7170 4859 390 5891 8189 3044 4498 6086 2447 8817 3663 7577 2155 7582 4930 615 1480 3794 7223 7982 7408 4990 7384 3018 5406 1474 3786 1506 4540 7036 8618 8543 1942 6840 9005 1697 3922 2230 2244 3634 1471 6890 7346 6922 2804 8631 3864 586 2064 6174 1985 4202 9054 494 1572 7106 3037 3289 387 5207 1169 7253 7716 6977 3526 7233 5872 3085 2937 1130 5338 418 8439 8259 3055 5577 6808 490 1189 2984 4974 5963 5914 5887 581 3380 8432 4994 414 3720 6047 274 8608 1608 7352 6028 5409 6049 3392 8220 6953 3183 2391 3618 8392 5452 2166 3363 9144 4039 9120 5765 6015 1317 4744 8348 3781 3271 8087 4760 4450 815 8571 8246 9057 5365 4423 7934 1469 1582 3027 552 3218 1046 504 1059 6152 2924 7546 2287 6309 4116 3216 7932 5749 1781 6627 8932 6233 7599 2360 36 293 7678 3615 2262 6467 1692 6716 4174 3081 1098 3600 2750 661 4705 9239 7680 2666 117 2475 3752 3508 717 2526 5485 7782 4795 8881 673 136 4153 7523 6738 8134 7578 432 8816 3624 6770 1900 7977 2677 5447 6073 8654 5951 6549 5600 751 5855 8950 4060 6090 5226 17 6435 2925 8242 5771 8073 3800 5787 536 5169 6194 7548 5695 4485 2713 2003 1650 5999 300 4284 3368 2737 7057 5204 7055 9105 1047 7576 4770 6140 7949 3813 2881 1885 2283 7160 7747 211 2564 153 7047 2910 1144 8157 4585 6275 8283 5173 4021 6918 6799 311 9022 2716 851 176 7373 3031 2552 6708 1061 3225 8 7086 5919 7448 5870 2407 8056 5816 4294 7978 7753 506 5522 5344 216 7886 4382 8430 2874 7113 1983 3237 974 3647 4724 8085 8441 2516 6513 635 2105 7958 7729 7024 8678 3015 6905 353 5214 1455 2184 2062 6157 4647 3524 3652 5431 3423 6539 3045 893 7129 212 3771 4678 8665 5241 5198 129 7800 2703 962 3965 144 7457 7749 2894 3287 371 3442 7695 8264 5321 5337 5732 5120 9109 979 4253 3126 2495 5434 6107 6582 3455 5206 7893 4910 4756 5258 6409 8863 8582 3830 5561 5862 1577 1719 6971 5860 6984 738 8521 3181 6871 5547 4209 5166 7569 719 1989 1648 8142 5354 8655 2363 1998 361 8715 242 4628 8287 7790 3674 879 960 699 4757 3388 2061 3509 3038 7331 4726 6992 7075 3692 7142 6889 4358 8322 6649 1409 5190 5491 6784 5723 2775 1067 7898 1288 3308 4295 424 2752 85 3777 4627 4158 4848 8795 6387 6583 1592 3247 9104 8064 48 7177 5331 9173 813 4825 281 3294 8440 7808 3703 1919 7358 2141 3835 7485 6588 8789 6426 6050 4718 1666 8928 5598 2275 7525 6603 6200 4112 8222 8996 6556 755 2724 5585 4180 2328 2424 2347 7468 9226 5768 1012 5277 8576 4864 905 1065 4196 7816 2735 5593 7514 2957 3790 6650 7168 9135 9012 936 6722 226 4591 7013 5798 2725 6921 3985 2025 6196 4725 5551 3686 716 3996 1975 1868 5133 3807 7639 2647 965 3307 7828 3428 7110 2624 7045 679 7589 6552 7417 5223 8302 6055 1044 1969 3215 1765 4876 8005 4411 3127 7439 1973 2705 7699 2798 7206 2503 4713 5317 6371 4665 325 8125 8281 2433 3329 6590 2680 4231 3935 7895 7626 5785 4145 4731 1283 839 753 5267 4815 6217 7414 964 550 1907 9055 33 2301 947 7031 5854 2206 4086 1532 4491 5457 6598 4443 4107 5182 5047 8415 1228 1816 8460 1815 2875 139 6267 1285 1000 5545 4844 6369 6548 8476 8483 7983 7040 3341 3912 3199 8096 643 426 8620 8390 4247 1640 9059 6946 8200 4074 1977 6711 409 2059 948 5002 9069 2662 8036 7300 7284 530 8635 9255 6672 1611 2954 4422 8331 1174 4640 1495 342 8059 2103 7667 4130 331 670 1339 5078 4652 188 5946 7076 3057 993 7306 6351 3797 464 5352 1800 570 2138 1490 5607 5845 3138 7183 7270 2092 2364 7587 2879 7701 7273 2975 5147 9116 3675 3641 4356 6251 2942 3506 6898 7736 4260 2916 1058 3259 8298 6187 2876 4316 4889 4694 3441 2692 4817 1331 125 3700 1990 7730 6338 2592 6191 3389 9202 9068 1428 2030 3848 7620 4927 1202 8596 976 4454 5987 7628 2935 7607 3146 3765 629 2213 9094 645 1209 2799 3688 4697 3580 4489 1036 4334 3451 1035 3906 2242 6665 5554 6526 526 8217 4693 3788 4942 2919 5815 8261 6580 5864 6692 2946 6332 6008 8956 3480 8316 8733 7231 7098 9238 4404 7410 3691 3196 7787 4553 8747 9294 2587 1844 7767 6037 2308 6820 6193 2863 1751 2174 304 3021 2646 613 1691 800 7974 3104 6518 3967 3850 3475 982 3457 6232 637 5455 515 1018 6876 7869 6061 4715 8364 1600 556 399 5009 7627 4537 2131 1430 3833 4777 3499 8094 7320 2529 1359 8844 8485 6981 2039 5213 2370 3900 5039 4373 555 2479 1579 8831 1071 6002 5079 3161 3100 5179 4675 2550 8659 5621 3555 6214 8918 491 4029 8952 4603 646 3454 2669 8694 209 2253 4179 1353 8290 6666 9270 1242 4028 9246 2903 5631 7495 5101 7555 1604 7742 5589 167 1048 6070 8239 4173 3358 8554 297 6536 7634 2744 2063 6909 6374 8109 1965 4676 8591 5254 6021 5625 6004 7382 5932 8983 2788 3402 5058 1672 2532 374 2102 6216 7770 8422 4836 5753 4622 4614 6756 2830 616 2761 7252 4662 3155 4787 1322 8293 9143 5848 1493 3059 8389 5925 5459 1615 5884 1842 3602 6664 1273 5663 2453 2854 4182 6422 5632 2101 8535 3879 6674 4514 8704 2146 6080 7381 1581 4532 8709 3071 4655 123 1905 5742 9056 5703 2402 3169 8913 6456 1327 8617 2721 6295 712 5662 1040 3429 9099 8120 2955 6356 8759 8304 2463 5296 4245 4312 8172 8705 8826 9293 3888 3016 5165 8818 38 2928 1827 7343 5871 8318 114 483 9174 2811 6656 2742 7403 6912 1350 5111 6283 8033 4810 1981 6815 1352 2383 3165 3714 1801 8004 1821 3324 4526 8630 2066 1661 181 7857 3587 8158 6686 3836 403 7961 8974 5745 9274 6793 2841 4993 6972 9312 1647 3970 2216 1908 7441 5303 3501 5654 8206 4673 6434 396 4860 1487 6801 3276 4626 3643 805 4352 6884 1370 1848 3612 4400 8920 5530 1660 4064 4881 1410 8355 3058 1645 2422 347 1390 3493 4926 4132 459 2202 7868 473 7666 709 2898 3281 3347 5738 718 2038 1980 102 7919 8936 4865 7275 7763 7572 4309 6293 3437 9140 6892 5454 2217 9211 7904 8197 7295 6268 7162 4346 1623 9124 3448 3497 3699 6685 4031 3019 4849 3905 204 4858 2625 8773 466 7249 7226 5163 5717 5108 6414 7180 389 3264 2067 1575 3005 3440 5417 6578 49 314 6796 6619 4658 130 872 3242 6139 4225 1115 488 30 6899 8900 1805 4115 1002 3312 8604 7071 8987 5308 9205 2702 1565 796 7325 2887 7088 7880 4728 1523 940 4401 5159 71 4166 1916 3486 7350 7962 7360 2144 6313 6690 4098 7979 1454 430 24 1521 240 6617 8335 443 3994 1530 7527 8515 4587 6538 2637 2627 6006 8634 8642 2886 2168 8273 2032 158 4947 2191 1656 4794 4612 786 468 2255 291 3353 5515 9114 5983 2642 564 571 8564 5194 7078 9097 9138 8766 6470 6563 5189 276 7336 7618 2474 2883 5153 29 481 5565 2988 2581 8860 5842 8719 959 6442 6564 439 7109 4938 4941 5693 1262 8810 3543 2170 7841 5896 6339 7675 7228 2849 2530 8324 8141 4466 4695 2594 727 4501 6134 3214 5469 4255 5114 4227 68 8629 9193 2684 1337 8606 3803 1001 4809 3894 4567 1355 622 5273 1094 4207 5093 3846 3844 3770 4981 9291 9036 160 1641 2580 5268 8211 3369 8020 5088 2766 5759 4912 1799 5046 4818 9092 6340 1013 6005 252 1548 4239 4933 4344 749 1871 5291 7993 978 694 161 6317 2208 5587 3226 7788 544 4959 1795 8203 283 3533 7255 649 4367 7334 4944 8269 3410 4393 5615 3241 8953 3195 3342 2787 1725 691 7413 7357 4960 7769 4251 8901 1481 5389 7124 3111 5825 2194 7741 8597 4751 203 5122 525 2846 2997 896 592 1142 4362 6189 977 697 6537 6565 5968 253 1737 5278 6547 551 4155 6696 3827 926 1347 4689 1732 6976 2816 8752 5668 9008 5445 5727 2976 8557 9227 3858 5205 1121 1367 6389 5875 6996 8399 7327 612 4186 2161 286 6865 3927 6540 6285 7751 3046 7480 5479 7323 7060 1593 3537 5993 4983 9083 2017 7112 4286 932 2736 8326 163 6272 8105 4318 1465 8975 7196 5570 1681 8232 9235 6925 27 7415 1218 5290 3837 5912 5500 6238 2632 7984 8146 2260 2528 6573 7887 6453 4313 5110 75 770 672 5399 8878 5797 7020 7957 1106 7309 5 8183 44 7765 3729 4332 3926 5702 189 5364 970 1057 1241 7805 4643 9047 43 728 1519 8556 5637 7712 607 4958 6791 6657 3172 3942 8703 2654 4248 793 2057 3514 5658 5685 3722 921 7536 3897 7030 4749 8993 3633 3492 6225 8371 5747 4606 116 9151 4345 5044 6391 2085 5634 4692 4123 7008 4470 529 3619 2173 7565 7574 3782 1210 5901 8162 2004 5433 346 2576 2209 8830 7784 7535 5616 1595 8121 8152 8467 1624 4956 2847 4125 2228 4162 7754 3434 2348 1399 2938 3573 2312 7330 1468 7073 7947 8077 5097 9229 5769 7184 6869 107 3068 7724 4442 5420 8912 5829 1651 3008 4088 7151 5439 7839 3828 9141 3151 6433 568 8849 4767 3694 6180 8905 5952 508 8717 8727 3249 7316 1267 2895 6933 2478 2225 8989 429 271 8856 8110 2823 6215 2436 6828 527 7161 6951 726 8282 5383 5208 1734 3532 7825 5518 2920 4395 2286 6759 6388 2634 869 4555 2534 1330 5339 4632 2660 2460 5623 562 8185 3750 994 108 5275 4109 2254 2412 1717 8850 268 4522 6242 8213 7575 4360 6814 7345 5440 9244 6779 2323 1 5460 4607 2097 2018 7050 2454 6652 6851 4043 8029 3217 4782 8562 6957 5878 4639 6753 1923 5370 911 6126 1229 8695 6628 5312 4376 833 6206 8207 8473 1689 2182 1639 7672 2302 2279 7227 1991 1665 6794 3405 4385 3513 5960 2356 5232 8965 1219 6812 2613 1245 6550 7506 3572 7891 6357 760 5018 8762 1030 6749 4342 5633 9266 7102 690 5257 1442 3515 357 7640 3977 4280 367 1877 2636 4055 7099 3198 7636 2956 8083 3010 3421 5692 5081 1489 2126 2603 3731 6775 5859 8274 3007 7894 7299 7328 4866 1598 10 6800 4918 8180 3756 2313 7473 757 5488 5740 3638 2531 1543 8314 1935 1404 9025 247 8893 126 7896 3424 9251 5931 5800 7488 3431 830 6057 7472 54 3576 6575 895 3759 5310 2204 7877 401 6219 1345 7293 4325 3110 630 5737 2336 8512 7702 7755 3982 7435 8092 7194 1377 6620 329 2970 6084 6383 882 5982 2620 6148 3220 4069 3585 6591 608 9216 8761 2760 4244 7187 4964 5481 4465 8076 4428 6171 6168 6527 2566 3861 5004 2149 8218 4761 7401 5542 3986 1515 4552 3534 9001 7497 860 7550 2966 4099 2299 8468 7247 5688 462 2343 999 1341 4906 3350 9212 1181 4599 2417 2520 569 7261 4831 576 2339 2852 1958 4206 1427 9218 4806 7044 2939 8327 8341 2056 199 8052 1282 3943 6341 1940 5533 2537 7396 6160 8014 192 7329 5644 6201 5040 2389 5238 6585 1887 2073 6903 9169 8080 4097 1601 9236 3384 3889 1190 1362 4139 9191 1825 4908 3505 8982 1172 6288 3373 2150 7925 3302 6471 6901 2130 588 7175 3014 1458 257 746 2882 482 3200 2016 4769 1132 3928 7090 9285 8677 3211 1500 6819 2180 2499 986 7356 7713 8214 4528 2022 938 4478 1447 2837 3519 5428 7059 3616 8943 5613 302 8113 3372 3383 8712 1128 5032 2824 3921 4330 3629 987 3222 6938 1688 2110 4903 7339 3011 1620 8574 9295 2831 1573 8391 7332 6058 8066 8593 5130 4663 5403 2979 8948 7874 8382 7349 8957 6176 2420 4741 2945 2051 5526 7938 5520 6155 8885 7903 1791 8815 2303 1045 3109 3036 8749 3143 7459 6296 8461 573 8321 8755 1501 5603 4601 7039 404 3569 5974 750 1566 5594 2400 8879 3639 8549 6521 1750 5281 3819 2291 4307 4096 5964 4433 1696 3240 6192 5574 1017 425 6075 8310 3769 1970 3662 4146 1166 2861 6786 8872 5675 723 1478 3723 4559 5070 8219 3432 5894 9004 7268 6902 47 3535 1864 618 4746 4078 4946 2160 7633 3693 5293 6136 1943 5221 4703 3764 3275 5653 2084 3899 7450 6468 6478 4714 6529 7056 4712 120 2635 9233 5667 4062 8552 2143 2050 6849 9247 9166 1170 2589 9253 7138 5873 5197 7376 7366 4126 8360 2649 1659 6320 4984 492 3607 8754 4913 2994 3541 1726 5941 9310 1585 4384 1722 8108 3746 5719 3995 8154 1502 2430 2996 4961 1727 1305 4733 3114 5077 5790 6709 7933 4536 8095 1343 3584 8495 5098 6850 7824 7265 4218 2118 4596 8690 9061 5995 677 8765 3159 5809 6257 6535 6853 7361 2445 7705 1978 5696 7671 8848 6324 7204 1161 4322 614 8470 1767 3880 4755 565 8329 1151 4321 3284 4774 6705 4299 410 731 3076 6112 8469 3166 845 5857 7693 7282 1397 8022 2547 877 4085 3717 3330 4191 3024 3253 610 5709 708 3170 5548 5438 308 5027 3597 6020 245 166 1952 8687 3310 2104 2462 776 9157 744 4875 2688 7772 79 118 739 7436 659 1782 4644 8969 6122 8757 3502 3079 4638 4418 6607 6359 479 3955 1178 4884 3345 2377 2655 3924 9033 3219 5100 8884 4340 2411 6386 3245 8909 7038 8405 6516 1155 3460 472 8138 1809 7676 4550 1024 5347 2278 4979 5367 8202 7809 5726 591 1257 1939 2513 1167 7719 5846 2467 4420 7230 8070 3069 6567 6846 8266 8112 8779 7219 4315 5911 7914 2539 6841 806 1472 5216 8245 7246 2772 8171 8404 8034 3084 9103 7256 3340 5392 9019 7778 3386 8349 2803 5311 4991 8558 7761 9030 8153 385 4183 3207 3966 6502 8031 3649 4051 5306 2700 91 3696 3238 1721 7855 7235 6546 3444 2923 7707 7470 8782 1743 6531 8814 7491 6253 1457 1393 1633 8774 7557 3054 7153 5938 3862 3049 8381 7644 2897 2749 4911 819 6834 62 9278 5225 7391 5979 8882 3102 484 4464 7089 8938 2472 763 4969 2414 471 5219 9132 4235 8748 8870 4873 2747 6955 5075 2119 4686 6224 5178 4554 1936 5437 5327 2973 3798 7097 3145 2541 8086 2459 8540 6051 2679 175 8488 6305 668 9136 3463 2985 7563 2223 4829 296 5243 7665 1301 2745 5511 2909 7581 1312 1287 6595 7023 7449 6864 2657 7625 1893 7335 1014 8063 575 475 5319 8636 937 6188 8433 9111 2810 5835 4600 3248 7593 2098 1021 5597 4525 972 3375 3527 9011 6830 4091 5516 6873 6104 7564 6259 7596 2549 2471 4790 5596 742 5444 6429 2638 729 6533 5318 1949 5244 3972 6349 3321 7804 961 7365 3772 3449 2517 2656 4897 2226 1711 3920 1137 7427 5145 4949 4458 9000 3876 8176 5711 2458 4902 5380 1686 3544 5436 1914 1437 6450 7959 8906 3601 3063 7764 2683 4087 1198 2268 832 5482 703 7016 6645 7203 7643 5908 2095 5672 8148 9321 359 4816 5345 1706 663 2183 137 8771 5865 2486 5105 323 6807 5351 4995 7931 1450 6247 664 6894 217 4967 6097 3137 4564 1606 8929 6412 2508 5082 2704 56 4574 9262 8581 1406 9298 4215 1733 946 3682 865 1203 2995 8366 7186 7120 7156 7462 2579 8822 6452 2011 3404 3231 1080 1898 7248 6424 1760 5680 6040 6466 8144 3250 2604 6345 2333 2629 5006 7006 133 4256 5697 7561 7821 2806 2187 8323 7132 722 1483 5813 2457 4040 1764 5836 602 8851 7844 8675 138 1724 3332 7901 715 6750 7257 4194 1996 4957 337 9023 2789 1383 8867 1175 3710 765 2469 306 246 6997 303 1929 582 842 5449 9225 3488 6959 3235 5083 6623 6745 8384 2719 3779 1839 6751 1118 1043 3 6401 1947 438 827 875 6913 8205 4265 6117 801 3482 112 924 7756 3795 3748 3154 1440 2298 9128 1655 3052 2960 5300 4242 2212 4855 2818 3521 2918 6300 6669 2072 8481 8605 2384 7250 9112 8931 8995 2990 5652 8013 8106 366 3090 1964 3339 734 4065 5026 2372 644 9080 422 4932 7879 7952 2321 5405 3776 7301 2544 2147 8603 4543 3056 3503 3687 6054 5372 2621 6737 5324 3477 7992 5052 2239 7432 5042 6505 7955 3744 4457 3255 4563 5376 5023 3034 8783 8208 2274 7423 5503 7375 3427 3297 4548 2771 9016 2940 1349 6254 3821 572 3993 6949 4054 866 8500 4867 8048 2496 3156 8352 1867 3997 6635 8671 9067 6377 4156 693 7150 6755 3595 543 2041 7185 6907 3545 3568 8479 732 7721 6149 5397 7081 3190 9013 1690 1069 8484 4605 3891 6803 8874 8718 3097 1272 6396 6597 7780 2031 134 3436 7792 397 6222 1930 4171 4016 9090 4792 772 2376 8438 5461 4719 3311 9101 3083 3089 3407 754 2164 3981 2941 1883 3467 3866 6917 6947 6532 5916 1569 7524 5886 3560 5808 7519 4113 1108 1957 7027 316 3139 6172 510 5264 9071 5410 2133 1889 5523 6282 6893 847 8883 1554 3439 1499 8992 724 3610 6874 5856 2612 9221 3823 4830 7315 3135 1683 1184 735 463 8260 5707 6523 7920 5620 662 7923 1710 383 8921 7815 4355 8785 1008 7537 2951 5142 7883 904 3257 8471 2236 7362 5849 4505 2574 1884 1963 2324 6312 6504 7340 5362 583 3608 3605 6525 148 5573 5418 9207 8852 2543 7280 593 3376 6115 3677 5033 3304 3504 3378 2280 7043 1429 3374 4338 170 8839 2906 7372 5928 3648 233 5694 2699 6237 9158 686 5699 3087 4494 4214 1588 8528 4504 990 5007 4608 3885 6554 1826 8559 1276 5230 559 7709 3975 9058 1891 3845 4722 8940 7127 2045 5743 5251 6510 1830 6439 4874 7504 3315 7164 4877 6395 6594 5509 4558 657 6281 547 1498 201 7980 6363 5112 7717 7867 7580 6304 9182 4834 2616 3751 8829 1182 638 3839 1289 4701 2152 6698 5065 8117 3060 2527 7471 7595 255 6743 8178 5966 9015 5126 9168 5861 9088 1186 6916 6694 7364 6477 7517 2172 8911 8024 8707 6760 7363 4136 2738 6114 1596 524 9127 1063 8286 1644 1344 9198 6961 9222 5708 4444 3236 150 1152 1340 5670 5821 9268 4750 2437 4066 3690 4717 8379 4569 6228 1783 5953 5510 6063 6624 3131 3919 4419 9292 2137 5701 9232 4124 4380 3105 8093 6052 8656 3784 234 6010 5948 4018 859 4406 3266 9252 8563 1704 4101 5885 1122 773 7943 9065 2754 5231 7954 8032 5463 1042 6250 2293 4002 4871 442 5921 2237 2327 8842 7830 520 3873 5170 6629 2678 287 7708 2554 7528 3838 2326 4683 7762 5462 6009 7338 231 3320 7319 3849 4736 2096 1862 3144 7272 340 6733 2694 8986 2873 8767 1290 8165 7989 5113 7467 777 8459 1636 4157 6569 5089 4895 3564 2922 5679 889 2385 1776 4317 6227 2545 7998 4988 1052 665 4783 8277 3337 2668 3668 9315 5793 8309 2958 6599 894 8489 7547 919 5971 223 3815 6495 5263 128 5138 4233 4437 8091 2950 5424 9209 5378 6417 4869 6960 5000 4905 6823 6062 7101 2405 2214 3274 5604 7911 1028 3962 3133 8587 2643 1253 5676 3064 9002 7021 6878 289 4424 934 6235 6292 179 3609 2899 540 6372 6033 6939 1528 3296 4945 8805 7819 6331 8480 2489 8173 523 93 8775 8683 5377 4868 239 5705 3742 667 5591 6274 6816 5060 3130 8250 4102 2153 5739 9121 142 4843 7292 5419 9188 431 6586 5906 5199 3811 8400 8128 8271 6718 8517 4210 7605 6910 5777 5099 1993 169 4276 9196 1845 3707 1658 5734 8530 4267 8465 1516 2380 6260 1618 3852 1904 7885 856 1542 9281 5017 2623 1865 912 1236 3872 3940 3180 8039 621 2122 8985 7835 1381 4128 7881 1216 7405 3934 3721 3117 1987 1255 2659 4633 2456 1376 7445 7313 6679 5446 8150 2317 8071 8735 7802 6185 5524 7438 1777 9123 6 5772 8301 8017 3344 4230 1131 3763 213 2015 9048 8100 4727 7409 1157 6159 5879 3789 154 3269 5562 5712 7496 9045 5575 1260 6835 9046 156 358 7902 1379 4223 3210 4266 7870 7144 6734 428 2378 6637 5143 373 2398 6817 4094 3184 4348 840 8580 4429 2282 2652 402 8295 2512 4880 4856 1009 8011 4598 5505 7479 5064 8104 1223 80 9194 7758 1522 6221 542 1682 7064 7594 5930 3944 6270 4075 867 1055 9307 5379 1466 8196 2978 2504 1037 4549 2028 4076 9063 4301 5801 3507 7355 2042 3168 1537 413 2037 8454 4089 8190 7464 7143 8616 7659 5146 8902 5401 5028 4232 8101 4007 8380 2558 5724 4542 9113 8362 2373 8377 7446 7645 5977 3956 4720 3496 5043 4371 2791 3086 3239 6511 2123 2982 9237 7214 6880 3999 1956 8167 5001 8346 6302 5280 6503 4377 6660 8164 1508 3516 678 3529 6198 8363 841 6630 997 5601 3806 5103 1590 2023 2362 1314 6868 7662 3002 4193 1097 4840 7752 2330 5592 8340 8672 2776 4592 5384 7568 444 945 5750 2306 1924 377 7108 7218 6631 6308 1313 7820 4745 5121 7686 4137 324 1740 8496 1147 4682 1435 1649 6870 5326 4305 1525 8917 8457 1881 5320 8057 9213 6146 1567 3860 487 4506 235 6560 1945 8660 9177 6230 7615 7813 8961 5710 9 2569 2893 2234 6042 8598 3408 3980 2220 5282 7337 5084 8643 7653 6128 5779 7694 4839 5493 7051 1019 1988 8529 7082 8448 2936 7606 1911 3433 3023 4921 4659 5358 3724 8935 5780 4037 808 6514 1982 3309 818 4853 362 3033 34 549 408 7003 7379 1129 4561 3586 8333 1836 5858 7731 8934 4030 3651 6197 368 7971 4519 4804 1875 2914 8573 4426 5922 3908 2361 2222 6978 8133 4824 6487 5167 2000 9086 5910 8958 1050 3658 7166 6781 5841 2727 2192 8809 3147 4551 1966 7383 1583 8235 5008 2210 6592 3792 2748 5080 5627 850 4189 1879 5767 1326 6730 2316 4732 8890 3637 6365 8161 7157 4667 1563 5224 9181 4364 8156 40 6762 8588 6286 3221 1497 9190 8663 6995 3824 1114 4228 435 597 7924 566 9066 1264 5687 2211 1358 8417 3809 1271 3391 5236 8513 4503 5340 7515 5645 4070 5605 8541 3494 5962 7397 4793 6162 301 9079 9034 2148 1995 1102 1657 7661 4394 5943 3758 1768 2929 6811 437 2392 3067 3946 706 3470 4044 6315 6774 7542 2394 8673 1847 4904 6987 447 1693 9041 185 6954 6034 2757 7033 4670 1384 2074 6626 6771 4917 8553 2014 4861 1213 5969 768 7657 5305 2585 3360 1915 2869 1275 4281 265 3478 2319 8651 1164 4417 4379 4509 3539 1328 97 109 2461 1380 5608 5536 9003 8212 5839 8332 2514 5157 517 6431 2431 4293 8160 9189 2189 4699 5049 7167 5304 5700 8444 5228 8464 7412 1746 6405 5155 8910 4459 2440 3481 12 7305 2205 2640 7972 7303 7481 7116 3072 3191 3282 4190 2501 9035 1627 9210 355 6725 8625 3870 6384 992 8903 7466 2354 9322 7585 5929 4142 6740 8692 5150 2139 7956 5830 477 6147 6449 967 3679 375 3522 8015 1180 8149 6766 5053 8191 4331 5398 1387 8681 441 5391 3728 7070 3426 6723 202 1227 9137 6425 2509 1698 6161 4095 5375 2553 901 5451 6415 6211 8895 1296 77 7179 2998 2987 943 6106 2583 6025 2325 5535 2608 3867 7374 5067 7380 6587 4290 4219 456 1779 546 1810 4119 6765 4081 8234 3192 1552 6436 4440 4020 501 6963 3630 5181 6515 826 7198 7245 2857 1544 6661 3252 7860 2753 759 8307 5893 1027 1375 4302 7556 2238 1797 152 4500 7530 3179 8254 5955 4773 5626 7451 2560 2294 6778 7308 7846 4476 2252 7970 8006 3115 8053 3254 640 601 5661 7210 5269 6094 6167 391 1158 6943 6131 1992 8025 5828 7425 692 2673 3998 5041 804 3706 514 747 7669 3073 6741 249 8236 5024 5913 1663 927 7647 6102 5996 3295 5571 1484 8586 9087 3933 5720 2598 8725 5135 8279 4971 3201 2784 7750 5109 3843 8038 6931 707 9164 9289 6858 8336 8963 5333 1088 1401 1453 2178 1304 1230 1372 1432 5334 4814 8584 5507 191 6027 604 5422 6676 2911 2176 3719 2710 8896 8763 6408 8493 2480 113 4436 5172 1560 3822 7266 5171 1556 1597 3775 3377 8151 7189 8691 7811 1177 5757 6542 3754 5965 6872 7579 1808 1092 6355 1937 855 6792 6170 6782 1735 330 7421 8868 836 3557 6068 983 838 849 752 6076 1876 2083 3299 7801 50 4642 6018 5762 1870 7965 1090 2590 6064 783 6746 1250 290 8041 6942 8345 7852 7169 1212 3826 5514 8058 5286 4439 8079 5949 4001 2340 8000 5336 7759 995 857 8657 6488 2614 6397 8888 3881 6461 676 6570 6985 4748 798 3732 6747 5640 5926 9217 5883 7697 8731 4887 5074 4375 8503 5590 285 4857 6589 6314 7431 557 4339 2822 5973 1239 2715 2331 9314 740 6334 2868 683 989 1299 2819 9272 814 4700 624 1470 7131 4679 4704 5959 531 7469 3747 7126 2778 7610 4241 5212 1015 5284 3973 3695 1557 2290 4059 7085 1041 7776 4547 6777 2826 4771 782 8385 4641 8504 7775 5294 2468 6043 4015 6179 6735 6555 6476 6783 1417 4934 7873 2425 433 2429 5168 2120 2387 3300 2112 7234 2675 6393 5799 5635 539 5066 6824 636 8720 3549 2196 8337 5188 6512 8229 2035 2510 2506 8216 7026 3355 7430 8637 3660 2069 8834 7147 8103 4163 4936 6443 1335 452 4208 8037 1913 3566 8612 647 3520 7005 8857 2396 553 8163 6888 8721 1614 2464 8919 278 9050 6928 2442 7810 6337 6024 1488 758 6246 6675 4475 5016 8610 2986 8693 5322 4188 656 7125 6241 8746 4254 1348 903 8065 3293 8443 9150 9192 6127 5192 3945 4681 7454 6455 2186 3129 2756 2266 3578 4289 2536 9224 4006 3904 2671 3088 2107 7104 4138 985 6316 6264 376 7500 5629 5373 9187 9122 2257 9301 1031 6277 3094 3290 6572 3538 8861 1925 8372 6615 2892 8406 6670 7935 2674 9219 2593 5984 5741 9319 7195 1994 2851 165 1662 37 8359 519 7942 1803 8937 2058 4390 184 2019 2709 4263 2482 1477 4680 5721 6243 6007 8174 1165 8285 1892 3712 3194 1300 1187 7608 8115 2969 46 7714 140 2965 2466 3288 2145 7118 594 7602 6543 2774 5260 785 4181 8247 9082 8027 3078 1407 4925 700 3091 7996 5494 9051 4413 2591 4108 3331 3893 5581 4127 5335 3666 1720 498 4430 5174 3932 2689 3907 8739 5802 8568 6118 6207 6640 58 4515 5539 6360 9129 1587 4950 9126 6491 4246 3469 2933 4602 8646 5528 1082 909 4323 1643 3322 9241 1899 6494 4452 3413 2338 1485 3265 8997 7592 3228 3570 5036 2833 5034 273 878 2322 1446 261 6208 797 2010 1016 7146 1792 7913 7487 8959 7458 6030 3753 8102 8365 6350 7490 6577 3562 5624 4216 3438 2917 6307 2926 3621 2667 3598 4878 6451 3485 6689 8947 2076 4711 3978 9062 4310 6911 6244 8711 1113 5246 8478 9032 843 7773 941 3136 2006 6887 4781 6614 6234 1096 1039 7518 3188 5115 6919 3594 2733 4279 5619 8955 9297 1780 1926 7529 3382 7385 3916 2860 4047 4595 6017 7122 1095 2645 7429 1730 4813 1192 5758 5253 7066 6742 5899 6166 1238 5125 4636 4224 509 3458 1389 2337 8062 45 5239 5537 8980 6534 5781 4306 6137 3990 5059 8398 8394 2388 5137 6096 8177 1917 7875 3367 1961 3286 1766 4063 4154 1812 5136 2773 2277 3627 5718 7851 1622 4529 7049 5820 4461 3542 2081 9183 4759 8609 5203 4819 603 7520 69 6790 1265 5458 3277 1127 2968 595 2739 2622 2329 6829 8374 6226 3092 2413 4758 4966 884 5961 6353 4573 5426 3359 4079 2959 4143 8238 3820 8855 5725 8251 9006 4837 1960 3113 3065 1278 7845 2198 2481 4374 5905 1880 6958 1828 8353 3039 2670 2793 7290 1671 5666 7967 6883 5085 427 1119 3279 6169 6780 1425 9206 1235 590 3476 1541 4952 8244 1669 8490 8621 3530 4968 6163 5795 1294 5806 8303 3856 3684 1078 6354 6847 942 3153 761 1613 2068 1226 8427 7000 3911 7460 4943 228 1784 737 5416 6915 2156 5773 7966 3655 8135 2502 2465 2551 458 1728 3713 364 5831 1136 55 5867 5117 5094 7511 6352 8542 4716 7722 8787 4036 908 332 7092 2240 609 5235 6867 3124 682 7326 6031 16 769 1625 1354 5472 1413 4763 3419 7011 3705 3913 3725 3709 4308 5671 1619 4421 1456 6579 2690 1517 669 7842 1712 955 2169 1786 8449 8916 7609 5538 1551 9290 6990 3122 7193 6736 1882 8926 110 6684 7864 6130 8760 6297 6855 7876 2639 658 478 3335 2855 7906 2426 5394 5552 1841 9029 7171 7288 3868 356 8386 476 9248 1922 868 9009 7229 907 5450 8431 3741 8780 7946 8769 3774 3162 2483 4409 7508 767 6493 8536 4992 5486 1274 7549 5330 6553 2615 5508 6026 8317 6440 5584 7095 7224 5698 5128 6721 5249 2179 7617 5162 7728 2351 1424 4820 1195 6653 4431 6576 3604 1256 4287 9283 1398 8169 4797 147 4530 7419 4303 8726 5360 5229 3414 3178 641 7212 2801 1463 1392 7159 3654 3118 2365 6069 1451 5880 2284 2334 6158 6680 221 1159 8595 4472 1123 6001 7493 6754 6204 5297 1667 2570 6944 9100 6129 5715 3896 232 1547 2650 2314 2800 1798 7202 3356 6844 5465 5141 4084 3121 1653 6545 4291 1225 1443 7818 7635 4203 1323 227 2794 7995 5553 436 9296 4083 407 3646 8184 4939 6145 6945 1254 4972 2043 4635 7843 2967 4336 7624 2113 6432 8192 870 8527 7539 1476 2345 7399 2902 9060 8375 6584 2584 9170 3631 348 1789 1423 6934 8258 5556 1240 5473 2446 7951 720 996 5677 3209 5643 415 4012 8526 3989 6906 3012 8923 74 3657 7452 7915 8572 127 5549 8865 795 3834 4463 7258 354 4589 3106 3787 275 5385 8289 505 6757 6860 2078 6982 7182 4008 4441 6773 312 3778 2127 2359 6151 8756 8887 197 5756 5527 8499 5612 4707 6459 4053 1896 6929 6441 1496 449 8626 5834 3882 2188 6732 577 4590 4104 5578 5847 178 2820 5443 2342 1513 6099 1320 2505 1819 9284 6065 6788 5185 7239 4588 6016 4170 1703 6464 3490 3381 1558 4033 5735 3070 7359 787 2318 885 8338 1787 7541 7463 5374 8716 5106 5580 5810 8522 3948 6731 4615 25 3447 4448 6821 4486 684 3689 1258 1329 3869 2697 3186 5736 8061 7948 8437 8658 7533 8358 6290 1756 8186 1011 6768 4987 3575 5038 7424 2596 220 3680 4314 1188 2406 1794 2730 5678 3062 6460 8486 2487 5942 4784 2167 1708 1076 4696 4565 812 5283 8408 8808 1145 639 4977 417 3857 1197 3244 254 7307 4975 5427 264 6078 499 1742 920 1555 3918 5902 789 195 2452 4176 2971 6156 1361 1208 2082 1252 7476 3635 313 1928 263 7727 8175 9084 6761 5055 4516 507 7134 980 448 2129 3548 4645 2878 653 8361 3298 7289 1574 2871 4738 6100 4222 360 6648 8114 3901 7297 3793 6482 6045 8497 2476 8475 831 8272 1221 6258 5211 1664 1514 9234 2247 8939 7823 7004 6764 7216 873 2932 6447 5567 2658 8140 8325 4462 4980 1610 648 1888 4365 2582 2758 4624 8753 1938 1171 2701 7287 2264 8575 984 7735 3263 1292 7291 59 3425 4575 7074 6279 6681 1863 1386 8233 4237 3233 180 9110 6022 1116 9010 7062 802 5353 336 5502 8003 1933 294 3193 2272 6641 8583 2159 4000 8824 4273 1374 4381 2077 1576 4772 8300 834 5270 966 1363 9318 106 7513 96 3366 1605 5217 5295 8145 1503 4148 1850 675 4524 2427 1284 5315 4621 5807 5936 3354 3683 7386 2256 4046 8253 2867 4890 8966 3232 2428 7433 1365 538 2571 6299 5673 284 8833 1866 3203 9145 6298 6704 3132 6604 3119 2366 1524 1621 6833 5250 5714 5731 2151 95 1205 8215 2432 6486 1749 461 7532 333 6663 8409 2665 7028 7853 6642 4901 2190 3175 4735 7988 3814 1824 771 6035 8859 3009 7368 7278 4914 2307 381 2792 8044 4630 2379 2865 174 6375 7370 5050 382 465 3243 9053 3343 7950 207 6979 3528 8045 7494 8555 4799 1146 1452 70 7296 5071 2439 81 4778 7854 5876 2963 2518 7733 7260 1778 1438 5950 7510 4785 1814 8846 6885 6378 4496 3854 1073 6810 57 7215 6896 8520 2297 5329 7456 1559 3656 9040 7660 4842 822 9024 8294 8978 9254 7371 7314 5704 6668 6837 2054 3968 2080 784 5850 2555 8072 2044 528 8118 6701 4698 8166 5316 1070 1580 4165 8538 3523 1173 4 2181 3984 5540 1897 7940 8397 5102 4434 3949 1224 548 512 1700 4197 7859 8994 5822 423 3303 6558 6390 8590 3390 6318 4022 2523 7739 7484 2090 3487 7674 7779 7917 4133 5939 1846 5791 9178 7597 1204 3150 5425 2494 8524 2434 1193 7613 5255 6048 8074 4955 4285 327 8099 3234 100 2158 7814 7105 1568 6074 2913 4833 563 6256 790 3590 5201 6023 1695 7406 8798 5649 2708 4791 923 1578 1491 2588 4576 4669 4989 3716 8599 8537 4812 7094 5920 1233 1373 7503 5760 7267 2633 8743 2687 8981 6335 1632 6924 4803 4144 4243 6520 9316 7394 2850 5784 3589 652 8210 2859 4508 1754 1332 6881 4320 3158 4100 1154 9256 4752 6046 4013 8954 1745 3785 4511 7181 6403 5684 1307 9117 4268 6263 7478 7798 537 467 3171 1694 886 2962 1859 5975 1319 9201 5602 824 6266 4582 2027 1034 4259 6445 9073 503 8050 9142 4161 3953 7831 7554 7342 3579 5261 1811 4080 1512 2371 3617 7663 5276 7734 8930 8669 5432 6634 3446 1068 6993 5944 2944 2964 269 4468 8155 6843 7649 522 8356 3577 1709 6085 2696 9271 7558 7631 1201 567 3452 1738 3950 6000 8429 5900 835 8075 8043 6528 6416 2408 6719 9276 1589 3465 8209 2779 5990 928 8914 9243 244 8811 9308 2099 2597 3951 1368 6077 6789 214 1110 1741 9131 1785 2249 6613 513 4004 6421 9223 6269 3991 7440 4909 3474 6968 2821 891 1984 4319 4090 5468 8187 73 4359 14 5559 7087 9179 2349 4497 7771 2759 4613 310 532 1333 3301 8777 4637 8127 8255 9180 2261 4891 6481 6248 5555 334 171 5648 6682 3766 3547 21 8737 5557 5770 5359 949 1388 8544 3614 1668 4238 6744 8908 8447 177 6724 7553 5195 7976 3273 3974 2116 5202 4517 6772 3472 3669 2896 4807 9260 9214 7656 6125 4024 5583 7990 2769 3963 7114 3812 6842 9115 6714 1026 695 2245 3877 3223 7888 320 1545 560 4493 5568 9263 5933 101 5196 6541 7737 8240 7398 3582 8506 4425 288 3313 6727 8412 4482 5897 8494 4688 2005 6608 4152 6825 8661 2231 6940 208 7377 7048 4534 6265 7155 5490 8018 9031 844 4324 4164 4800 5651 4730 8124 6469 131 6287 8862 5240 3767 4217 2834 115 3887 2246 4520 9037 8713 4296 4775 5903 4577 1263 7987 4073 250 1855 451 8624 6381 1527 3701 5279 8292 7968 864 7591 1520 485 1807 26 4205 743 8012 1279 725 6625 4283 103 4311 5467 6797 6530 7242 3352 3093 352 6861 4172 157 155 3915 5907 9085 8676 9052 243 3670 2767 4159 4455 4292 8770 7474 5045 671 2764 2511 279 1004 4789 1607 3082 3644 5665 5139 351 5805 4649 6964 5382 20 788 5421 2358 8129 8089 5227 5531 2664 5958 7768 5314 2395 874 2029 3017 4835 3780 9133 8907 6177 5957 6306 9230 9172 853 2309 4776 9154 7277 6827 4187 7690 7190 121 3917 8501 3417 151 4341 8130 1948 236 1507 930 6787 7601 8221 8858 1906 2858 8864 4997 1912 2786 2367 2808 6428 2991 5158 600 696 7240 7388 3718 5546 7783 1817 766 9165 6739 2311 2276 6974 1886 2448 7567 8744 5681 5341 4343 7760 2224 1371 4140 6879 2992 3859 8047 5622 9028 7822 2740 1266 7149 8570 2335 3035 871 8308 6941 6561 2270 3483 7442 6012 6566 7407 5271 8730 7745 6014 8688 2109 6029 3406 910 8119 5048 7999 5332 939 7725 8147 8265 9007 3357 3676 8136 4383 8474 76 3229 6301 3258 3929 5844 206 1752 2444 13 8734 6032 6932 4111 8204 8357 5309 4929 7715 9171 9228 3134 2368 4329 3227 6121 248 338 1135 5890 9199 292 5682 6262 1420 8821 6804 2079 1251 3743 8116 3306 309 5393 6110 2672 7682 5390 8652 2048 5927 2606 6419 3128 222 881 3667 3272 8837 7544 5483 8823 6712 5012 5774 8268 1561 558 1010 3567 2912 7263 5265 2484 7621 3884 8199 378 7766 2409 6859 3841 8972 187 7096 8168 7002 8925 596 2780 183 5869 5840 8960 2795 143 973 8023 7387 5233 6119 1584 8915 3736 8550 3498 8060 1459 2491 5980 7700 6822 5967 7091 633 4888 7392 2663 6366 4416 587 7566 988 9313 5131 3685 84 3292 8988 6557 6605 8122 1831 1364 8007 1707 931 8097 7650 6343 8082 1156 18 319 3471 5544 5357 2827 4541 99 5618 3445 887 149 3653 6485 2352 4651 345 5369 3205 4011 2630 4740 5827 2947 7264 9176 9286 1291 3531 7794 5020 3546 1680 370 2864 8107 7130 3325 3804 8009 5021 5513 4850 2519 1675 1111 3847 7929 7827 6729 958 8098 5466 1378 933 7909 4017 3030 511 6448 714 5610 4892 132 4106 5480 2134 2815 8305 1968 5395 3041 5794 914 3256 7796 2423 3808 2357 1616 535 2930 5019 7035 3074 4899 380 3029 3103 8263 4593 8042 4488 8945 7145 5895 6900 3938 2125 1494 4220 8877 4982 8632 5484 7865 963 4668 6465 7208 2980 6818 6184 7603 6475 5061 4499 6798 5733 4366 7706 4953 698 1927 6019 7878 7629 8446 4973 1955 3606 8602 4729 4412 4999 4687 9287 3626 7939 5313 8944 1788 4056 4052 1133 5669 5609 4354 3969 6165 4580 4067 1357 2829 457 9265 7637 4581 2601 5429 5187 6986 1602 5991 1005 8270 5566 4117 1553 6336 7812 4832 5978 3022 2369 9078 2777 6611 1141 4623 7777 9195 7944 7973 7806 9279 681 6517 5935 8280 7019 8262 5744 5387 6522 8689 3164 6989 4446 3886 4677 5245 4003 2193 2981 2108 9039 3551 1531 6769 9148 3592 5691 1550 6593 3152 2490 2525 6838 6474 4924 6203 5918 6108 8941 2136 7907 2853 4403 5985 9134 8198 3565 89 1873 1351 1534 3810 7201 6013 3025 8001 6697 63 1634 4611 1163 2521 3678 5124 2556 2838 4654 1626 6003 1979 2884 5657 2686 7679 6616 3316 756 1199 5118 1309 5506 3735 1436 3453 6500 9257 7748 6053 6677 2974 3174 4788 3540 6236 3285 4149 7140 8426 7786 8569 606 4326 5776 8686 6839 7889 6240 4103 2743 3462 9231 1054 4768 8946 168 748 7133 7502 82 1032 5161 7444 6141 4304 8653 2273 4940 7367 8804 2154 1211 5752 4811 7286 2060 5272 2498 6142 1346 4684 6930 9020 4560 7651 1526 3554 6795 4931 6066 4234 7817 6430 1673 741 578 4666 8257 200 4998 6683 4521 8296 8619 4618 8297 5832 8892 7698 1310 5218 8436 1473 2723 3461 2989 5945 7905 8393 3681 2856 1125 6098 2628 9242 315 3385 6671 2374 3941 7354 5184 6120 2698 5435 7573 5441 3189 3558 5599 4512 6370 1394 2088 7673 6802 87 2685 3336 6638 6072 4609 5292 1769 6758 6116 5877 3628 5823 4471 3066 1755 4372 8466 8835 2201 5252 4883 6398 1753 2065 1953 8028 6831 3730 5639 5003 1464 5989 6895 1183 2403 3959 3202 5396 8508 2353 6991 8442 5014 7791 971 6173 7785 6805 2522 5144 486 5148 2921 1303 3952 4150 328 1261 1504 3061 241 4808 4634 7174 8143 6490 4584 6089 2809 1539 2281 5411 9258 88 9119 1395 8840 3987 2052 3583 807 541 469 2375 3563 8507 2993 6319 1356 386 8768 4616 837 2171 8055 3976 7584 2943 4604 8699 2814 6325 3270 3349 8299 2605 4213 1628 4045 6937 2034 3930 4739 4764 3125 5072 1302 4178 4483 2731 5728 4617 3116 918 2825 8886 1179 4378 7863 4915 205 8534 9077 3176 6923 2682 5476 1222 9311 7882 3875 8970 5164 1860 2515 3048 2562 4920 5366 8984 5489 3697 4110 3878 4023 1316 7137 8344 925 124 7067 2128 7526 1878 1510 7053 599 9299 4723 1642 1248 3863 4754 4353 6399 730 3394 369 1652 8967 6935 4297 7046 2927 4057 3173 5288 146 1731 4226 6095 4041 4175 2844 5572 4477 7630 5186 1759 4544 6195 4261 781 3305 1134 1445 1806 1729 6970 3979 711 1475 1901 2741 2706 1910 6610 899 2195 8871 1298 8227 6687 3645 5092 7188 4408 8764 1715 4269 1391 1434 7158 7274 9130 8611 7063 2175 2033 6144 5156 4821 1033 8578 6092 821 4823 1366 9070 4201 1400 182 5499 7836 1433 4487 1419 1321 322 5614 7404 7238 7475 1903 858 7531 8455 1670 5030 2768 8491 8801 3101 3141 2251 8368 8030 159 2535 9306 623 1006 950 1148 4796 3553 3230 1104 6715 8845 393 585 2835 7220 4571 8195 8793 816 6249 8347 4027 1954 3511 8533 7176 2300 3947 3818 1449 295 8414 1207 7590 145 2983 8532 3831 3013 2477 7799 3874 1872 6079 5011 3318 460 1714 5154 1405 7681 951 4212 7652 2289 5453 704 6342 2346 2653 3599 2215 3829 4862 5751 4282 8276 6908 6463 343 7570 1586 8843 2418 3468 5504 3435 5852 2386 8418 42 4610 3099 8797 5037 8342 5464 6688 1959 9075 5057 2802 848 7616 4105 6245 8284 4200 493 1280 5215 2573 3937 7684 764 4965 7093 3574 2341 3334 632 4391 1835 3636 5301 8194 8638 6857 2695 502 3596 2842 1295 6936 6410 6327 7969 8350 1360 299 1444 3632 6703 7797 2285 104 8510 5853 8745 5022 7211 7217 4660 3466 1103 7677 3251 8492 2381 4646 7507 9261 4648 7632 7619 3761 1820 1630 3283 5127 1713 1818 1771 5015 7353 7068 2908 7207 1654 8078 1687 5521 4578 4562 318 8897 2524 4456 3484 5988 3704 2142 2053 5266 863 7743 7926 2681 1678 1951 3327 5796 3028 8668 1269 8647 8594 4702 4922 791 210 642 6109 5408 5586 5981 8275 6462 8976 8019 8644 2497 5062 6294 7683 2229 5788 9267 1570 8639 8516 1635 4798 5478 4058 1492 5175 6952 349 1460 2 518 9042 619 5237 5628 2907 4298 7072 2765 5400 5543 1081 7922 4335 455 4072 266 4438 4177 8614 4414 2441 7850 7015 5152 6231 8803 8565 7492 3760 4691 5646 7916 500 1629 321 5346 5328 8369 2265 2197 7994 3791 1971 5095 4586 2891 1909 2177 8904 6186 9064 6373 3817 956 4204 5650 6600 9125 446 1231 6212 6483 7148 3805 8776 7623 2271 4896 1150 7437 5683 5617 5063 4257 6261 5342 2421 2040 922 6559 4893 5564 1834 5746 5456 6776 6673 9200 8628 5471 2661 6643 1684 7534 4597 4034 545 7571 6278 3726 6927 3762 8139 5363 6060 799 3883 7121 6659 5448 9302 8081 5361 3672 8701 7498 1334 3983 6276 5892 654 7191 775 5234 8313 3871 2185 6544 2611 6311 902 1861 7732 2755 1099 6038 7369 2012 8819 7428 8633 4082 9118 5904 890 6210 8820 6213 2714 823 6508 3671 7103 981 6138 31 1723 5934 3898 852 8054 1220 1790 7588 4042 7641 5638 2540 4801 5289 1529 41 3006 6418 8463 666 5056 6956 7910 1315 8942 4566 5262 3142 554 2315 809 1594 1029 6056 2577 4985 6492 5343 6153 7612 6826 6454 4240 5729 4523 4405 7509 7658 277 2135 2087 3246 7738 2889 4199 8434 7847 1259 372 3702 4986 3740 1025 1793 2114 5086 4518 1869 1479 1382 7395 8577 6367 5868 7953 5940 7858 5010 1246 6658 7912 7209 1416 4410 7426 2344 6713 2746 4236 6071 7837 5588 4068 954 7009 4151 8450 7562 1007 9250 8084 7241 736 5068 5180 3698 6767 6691 1003 2443 5818 8640 3992 7711 5274 9163 5512 6036 1932 561 4389 8679 7205 605 2288 6457 267 3708 2885 64 4050 7262 7304 5349 3333 829 8378 7317 1999 8778 7918 6568 6199 7012 8026 9156 6190 2732 3768 5606 4252 3623 7687 5498 7243 7803 4822 953 688 7393 8648 8425 1196 3422 4879 5937 4333 8866 5302 8343 3160 4854 6271 8482 1538 8786 7642 2001 655 3278 4923 6380 8589 6950 224 8225 6059 1851 2117 8458 7271 162 4274 164 4709 5200 1107 1232 8751 8423 3095 8601 2770 4350 8740 11 3500 4583 957 913 7443 5764 4900 6181 1089 9152 6700 8008 6218 251 1206 3047 6124 4415 445 7757 660 1536 9081 9269 3909 1325 2263 5558 7455 5824 7341 3051 8188 6891 4907 5519 8278 4398 9106 2817 2241 489 897 7963 2872 7543 256 1461 9098 3783 5686 5655 4396 3903 1079 7991 7152 7400 4392 8566 6438 2295 5191 817 3387 5763 230 2091 2848 6385 1679 1091 7029 2221 998 2691 8456 5501 3964 61 5811 9159 1617 2258 2485 5833 5298 906 4546 7807 1921 4479 2595 1249 3620 861 3865 7416 1511 98 6877 6863 6904 8451 4357 4014 1823 2399 8796 2548 8462 7010 7061 5129 8354 6087 7670 5998 4369 8132 9074 1200 6473 4846 4742 2227 8827 2905 969 5730 7832 3148 6897 2828 1609 733 4685 1293 4025 7501 2845 7696 8894 9304 5442 1856 5368 8889 1426 8933 6809 5778 3379 915 5381 6980 1176 3512 3910 3291 6662 4970 2870 8411 2310 4432 3588 5838 1286 4802 8702 5954 7018 3902 7197 8068 5804 1705 4851 1997 8561 9264 4852 4625 2610 3430 3107 2781 2722 9153 8792 2218 8545 400 6280 1086 7119 8560 2259 2304 8424 7664 4262 2839 2607 5402 1770 119 6489 1112 7001 5851 5104 811 6321 1739 6239 2021 6154 4397 2507 5487 363 5970 5140 350 6326 51 6323 6852 2949 7985 1535 5775 3711 3412 4122 4048 6183 1277 7077 8179 5782 6646 8010 7849 9093 3096 3140 5087 3739 7512 6574 5176 3456 917 9155 5819 2866 8670 194 6446 3925 111 6392 4786 8435 190 1796 6519 6914 1822 2500 8825 8126 6706 4327 3960 3914 1217 3398 7200 3642 7646 6507 5160 8880 4826 6693 952 3206 9184 3319 3550 651 1153 5149 3737 8088 1049 3988 6717 7997 4480 7163 5882 6988 7746 3853 8315 4484 6041 1631 4710 6509 3801 7037 1270 6178 8551 1268 1085 7986 9072 1191 3749 7135 92 8607 4445 1467 3581 7848 8784 4954 9162 1829 8791 3280 8832 8924 1297 2641 1101 4531 7928 4572 8230 2233 8509 2235 888 8021 229 83 2782 4349 4092 1402 8237 6209 6969 453 2843 6581 9220 780 4427 8758 6289 5193 516 3591 6633 5630 6333 4300 2783 9273 5414 2728 3640 3149 3517 2157 1974 3954 8413 4118 2390 1403 5660 5582 5356 4026 5025 9102 2046 6480 6606 7276 9043 5992 2631 4185 3650 6083 8623 496 6678 713 3552 883 1412 8111 7192 238 2586 2071 3197 8742 2404 2296 9076 2726 5716 8502 7774 258 2796 3224 4898 8971 6105 9208 3755 3734 3177 3409 6501 2232 6101 580 4272 8724 2162 4721 6854 7899 9014 60 9259 1482 7141 1072 3971 2785 2836 5177 745 533 8615 7655 935 8684 434 8641 198 7025 3825 7065 8700 4388 5123 2915 3450 1306 6111 280 3816 6229 6994 7022 8407 32 5888 3346 846 7007 7692 7861 6150 6164 4147 2888 8328 3026 4275 8802 282 3262 1685 8964 8319 5407 6651 2934 5404 4863 2862 8428 1140 4386 8035 1944 1967 2332 8728 3796 6437 5783 2305 3323 2401 689 6636 3050 5323 2561 2563 2036 305 1701 7042 1234 7838 1736 8741 22 5947 497 196 3892 8472 8732 7390 4650 9305 5475 5497 4619 2676 6720 6182 8312 1214 2199 19 8697 7154 6948 7744 3268 4996 6983 4805 8800 3939 8402 8367 1415 5786 8854 1022 1849 5116 4490 7052 6639 5496 7516 9240 7945 9215 1832 7941 4570 8706 5247 90 8738 8016 2163 8622 5259 3370 9175 3004 3396 8487 6205 3738 8650 8680 2002 7351 7600 6413 4447 5579 3479 5837 1064 8812 7115 4347 4510 6252 627 4827 598 7222 6498 52 4264 6223 2538 1838 3267 3080 9303 2972 7560 5898 4631 3401 4129 3365 6402 7283 5299 6407 2320 4473 5889 474 7691 3673 8170 7225 2533 2711 9275 2009 2350 9091 4337 8876 4674 7964 4928 7100 8685 5817 454 2707 8729 8922 2415 6571 7483 8548 4114 3443 5076 1972 4387 3802 3661 1895 8627 6423 6647 341 8547 3727 8420 7434 6965 6726 2013 7221 7254 260 2961 4402 1852 5803 5492 6291 3371 5917 3364 7921 6202 1564 7505 6484 2977 5529 2165 6394 8899 8525 3923 6695 5090 8710 5563 7833 4370 1505 339 5909 820 6082 270 6920 2618 4277 7726 680 2397 3053 8841 898 6848 8714 1308 6702 5713 626 8649 4671 3518 8395 8123 2115 2542 3212 8256 3317 6458 9026 7704 6427 7586 6044 916 4513 5412 1448 6728 5541 4134 6845 5881 8579 7648 8973 1762 6551 5722 7499 186 8664 4828 5923 5576 7402 307 9167 8373 9107 7251 420 3098 8046 2720 1185 9139 8585 8666 5470 8370 5256 7348 7333 6962 3328 3261 7862 862 5073 3840 3020 1084 4838 1281 4363 7829 6273 4474 5525 6143 172 2952 344 7139 3625 2121 2599 7900 6322 6303 1411 6358 4916 1890 9049 5656 4886 1023 416 3561 5647 6364 7981 1338 5674 1149 2877 1813 7545 1518 4872 8243 65 1062 4351 1318 2203 7622 8403 7069 3733 7259 5866 3395 6806 193 7311 7604 8567 7703 7486 8998 650 1117 7318 6524 8320 5430 4160 6622 3362 1699 3611 8999 3745 1336 1109 5997 3931 7378 4535 6133 1833 7930 2055 6132 2007 3459 3000 6763 3961 778 2435 8090 4937 7975 7128 1124 1747 8990 3185 5287 7890 365 6832 7871 5013 4093 2609 1138 6420 1414 7937 4963 5863 2565 5532 9204 1853 7165 2762 2718 141 7461 5091 6091 8662 4557 5761 5748 384 8477 9309 4469 419 3556 4071 1237 2267 6926 8267 105 7668 1637 4361 6135 7793 3495 317 1075 8722 3314 7723 8201 1854 4460 1775 1837 8511 2880 3397 5285 7840 6330 810 5222 5642 9317 53 6499 2393 1546 7795 6362 4221 2075 7465 5035 2100 3832 8223 5956 6562 7482 8696 8387 991 4664 2250 7310 5636 6088 3491 4038 5183 2568 8698 589 9288 3603 5005 2492 2243 7422 3400 5915 5051 1763 3418 398 4453 3593 5415 6344 4762 6966 6382 5386 685 6081 9282 1950 4167 4278 8248 2578 7638 5151 6836 2470 4271 6011 495 5690 8453 6175 7892 94 7866 8898 7123 2488 7178 8519 5789 5560 6310 1562 1549 7312 8228 8396 8991 1053 2450 4779 1311 1051 4661 4533 1677 4894 3075 2493 9185 3187 2832 3665 1215 66 7453 1509 6400 7411 7083 2292 8523 702 3895 4495 4935 6103 7084 2200 611 8306 470 1894 5495 7856 8249 67 5595 440 2546 6496 7710 5659 1369 2473 5994 1100 1422 892 6752 3936 3112 1976 8674 2132 4270 8806 7789 6875 4250 4032 4976 5132 1902 2807 2008 4919 1744 8853 9108 5119 8505 7111 929 3338 8445 3773 762 6654 122 4753 7244 1162 6813 1247 8723 3361 4545 8416 4870 6655 2600 2617 7041 395 6785 4435 8069 406 8592 2111 794 3123 7418 7236 5812 1646 2559 7720 9018 6406 4121 1931 5924 2219 7034 7281 9320 5388 1962 1920 8388 1774 8546 2455 8334 625 7117 3489 4629 534 8979 8231 2693 8807 2840 721 1244 2890 6472 4407 7054 7834 705 3108 1441 2751 7302 880 2140 8962 8040 5423 8873 2093 9277 9249 7960 7285 4962 1946 944 4019 6862 7718 8836 225 237 1918 8847 828 8193 4131 4882 3040 1716 392 4005 6255 8539 3715 2020 3613 8311 1599 4766 687 5413 5972 5976 1674 480 792 5754 5325 9147 5210 2089 8927 2248 8790 1093 1074 5069 4527 1757 411 1168 6856 1773 1941 3799 5641 7872 8351 8799 1702 2572 3664 3536 2813 5096 854 1533 219 8600 5843 5534 9160 86 3348 2953 8049 3043 2438 4168 6975 4539 3032 7173 779 6866 6667 2619 6347 6368 7614 3042 4467 4734 3001 1342 7279 5706 4035 5611 5107 6361 39 8376 3890 8051 1077 2648 4120 6328 7014 7540 8419 6404 2999 1431 8869 15 7489 2355 803 3260 4978 5031 7344 8772 3415 5248 4449 2805 2557 405 8383 6618 4009 4620 2124 6644 3622 6123 3003 4579 1934 8241 3957 8252 6497 8828 7897 2602 2901 4951 3851 8291 1540 6379 1066 7521 5792 2070 8781 2729 6973 4765 9095 218 8339 4249 4594 975 8159 5029 5874 4229 1396 7611 7420 8613 8968 6506 7654 2047 1038 5474 6596 4653 4568 1591 35 6329 1986 8330 4556 8401 8498 6113 7136 3416 9027 8131 3411 6346 774 2763 8226 1486 5477 215 6376 2416 3571 1087 4948 421 3167 394 4657 3473 574 7522 8875 7058 7826 7685 2931 7017 7559 2382 3958 7172 298 1385 4690 5755 4492 2904 1874 2026 7032 5826 4481 6748 5664 1804 9146 2567 3393 4672 2086 2419 701 2651 4198 1105 78 4141 1748 7213 3326 8682 1718 3403 8421 7321 7884 1761 5054 7294 6998 3351 4841 4885 5348 521 4061 4077 8067 674 6632 7552 9203 7298 579 6067 1802 72 2812 1243 6967 259 6882 6699 173 1638 8949 7 1857 968 23 4328 876 2449 1612 2948 326 4507 3559 4010 3182 7927 5569 8891 3842 4258 9300 2024 5766 7688 9245 3855 5689 617')
```

```python
# sseq

def sseq(s, t):
    location = []
    s_index = 0

    for nuc in t:
        while s_index < len(s) and s[s_index] != nuc:
            s_index += 1
        location.append(s_index + 1)
        s_index += 1

    location = " ".join(map(str, location))
    return location


def final(fasta_file):
    input = [str(record.seq) for record in SeqIO.parse(fasta_file, "fasta")]
    s = input[0]
    t = input[1]

    return sseq (s,t)


fasta_file = "rosalind_sseq.txt"
print(final(fasta_file))
```

```python
# lcsq
def lcsq(s, t):
    dp = [[0] * (len(t) + 1) for _ in range(len(s) + 1)]

    for i in range(1, len(s) + 1):
        for j in range(1, len(t) + 1):
            if s[i - 1] == t[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    i, j = len(s), len(t)
    lcs_seq = []
    while i > 0 and j > 0:
        if s[i - 1] == t[j - 1]:
            lcs_seq.append(s[i - 1])
            i -= 1
            j -= 1
        elif dp[i - 1][j] > dp[i][j - 1]:
            i -= 1
        else:
            j -= 1

    return ''.join(reversed(lcs_seq))


def final(fasta_file):
    input = [str(record.seq) for record in SeqIO.parse(fasta_file, "fasta")]
    s = input[0]
    t = input[1]

    return lcsq(s,t)


fasta_file = "rosalind_lcsq.txt"
print(final(fasta_file))
```

```python
# edit

def edit(s, t):
    alignments = pairwise2.align.globalms(s, t, 0, -1, -1, -1)
    best_alignment = alignments[0]
    aligned_s, aligned_t, score, start, end = best_alignment

    edit = sum(1 for aa1, aa2 in zip(aligned_s, aligned_t) if aa1 != aa2)

    return edit


file1 = "rosalind_edit.txt"
input = [str(record.seq) for record in SeqIO.parse(file1, "fasta")]
s = input[0]
t = input[1]
 
print(edit(s,t))
```

```python
#countMismatches

def countMismatches(s1, s2):
    mismatches = 0
    for aa1, aa2 in zip(s1, s2):
      if aa1 != aa2 and (aa1 != "-" and aa2 != "-"):
          mismatches += 1

    return mismatches

def fasta(file1, file2):
  s1 = "".join(str(record.seq) for record in SeqIO.parse(file1, "fasta"))
  s2 = "".join(str(record.seq) for record in SeqIO.parse(file2, "fasta"))
  alignments = pairwise2.align.globalxx(s1, s2)
  aligned_seq1 = alignments[0][0]
  aligned_seq2 = alignments[0][1]

  return countMismatches(aligned_seq1,aligned_seq2)

file1 = "IL12A.fasta"
file2 = "IL12B.fasta"
print(fasta(file1,file2)) 
```

```python
#countGapOpens

def countGapOpens(s1,s2):
  opens = 0
  gap_opens = False

  for aa1, aa2 in zip(s1,s2):
    if aa1 == "-" or aa2 == "-":
      if not gap_opens:
        opens+=1
        gap_opens = True
    else:
      gap_opens = False

  return opens


def final(s1,s2):
  s1 = "".join(str(record.seq) for record in SeqIO.parse(file1, "fasta"))
  s2 = "".join(str(record.seq) for record in SeqIO.parse(file2, "fasta"))
  alignments = pairwise2.align.globalxx(s1, s2)
  aligned_seq1 = alignments[0][0]
  aligned_seq2 = alignments[0][1]

  return countGapOpens(aligned_seq1, aligned_seq2)

file1 = "IL12A.fasta"
file2 = "IL12B.fasta"
print(final(file1,file2))
```

```python
# countGapExtensions(s1, s2)

def countGapExtensions(s1, s2):
    gap_extensions = 0
    gap_ext = False
    for base1, base2 in zip(s1,s2):
        if base1 == '-' or base2 == '-':
            if gap_ext:
                gap_extensions += 1
            gap_ext = True
        else:
            gap_ext = False

    return gap_extensions



def fasta(s1,s2):
  s1 = "".join(str(record.seq) for record in SeqIO.parse(file1, "fasta"))
  s2 = "".join(str(record.seq) for record in SeqIO.parse(file2, "fasta"))
  alignments = pairwise2.align.globalxx(s1, s2)
  aligned_seq1, aligned_seq2 = alignments[0][:2]

  return countGapExtensions(aligned_seq1, aligned_seq2)


file1 = "IL12A.fasta"
file2 = "IL12B.fasta"
print(fasta(file1,file2))
```
